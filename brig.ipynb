```python
def tr(s):
    try:
        a = int(s)
        if a == 111111111111111111:
            a = -200
        elif a == 9999:
            a = -201
        elif a > 3500:
            a = -202
        return a
    except:
        return -190
```

```python
df["ID Парковки"] = df["ID Парковки"].apply(tr)
```
