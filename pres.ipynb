```python
import pandas as pd
import numpy as np
df_brigada = pd.read_csv("brigada.csv", dtype={"Исполнитель": object})
df_inspec = pd.read_csv("inspec.csv", dtype={"Исполнитель": object})
```

```python
def tr(s):
    try:
        a = int(s)
        if a == 111111111111111111:
            a = False
        elif a == 9999:
            a = False
        elif a > 3500:
            a = False
        return bool(a)
    except:
        return False
```

# Сомнительные ID Парковка

## Ответы бригады

```python
df_brigada[df_brigada["ID Парковки"].apply(tr) == False]
```

## Ответы инспекция

```python
df_inspec[df_inspec["ID Парковки"].apply(tr) == False]
```

# Валидные координаты среди строчек с неправильными ID парковки

```python
def tr(s):
    try:
        a = int(s)
        if a == 111111111111111111:
            a = -200
        elif a == 9999:
            a = -201
        elif a > 3500:
            a = -202
        return a
    except:
        return -190
df_inspec["ID Парковки"] = df_inspec["ID Парковки"].apply(tr)
df_brigada["ID Парковки"] = df_brigada["ID Парковки"].apply(tr)
```

```python
list_brigad_must_valid = []
for i in df_brigada[df_brigada["ID Парковки"] < 0]["Координата"]:
    target = df_brigada[df_brigada["ID Парковки"] > -1]
    l = len(target[target["Координата"] == i])
    if l > 0:
        if i not in list_brigad_must_valid:
            list_brigad_must_valid += [i]
list_inspec_must_valid = []
for i in df_inspec[df_inspec["ID Парковки"] < 0]["Координата (правка)"]:
    target = df_inspec[df_inspec["ID Парковки"] > -1]
    l = len(target[target["Координата (правка)"] == i])
    if l > 0:
        if i not in list_inspec_must_valid:
            list_inspec_must_valid += [i]
```

## Таблица совпадений с валидными точками - ответы бригады
Для удобства отсортировано по координатам<br>
ID Парковки < 0 значит, что указан невалидный id

```python
df_brigada[df_brigada["Координата"].isin(list_brigad_must_valid)].sort_values("Координата")
```

## Таблица совпадений с валидными точками - ответы инспекции

```python
df_inspec[df_inspec["Координата (правка)"].isin(list_inspec_must_valid)].sort_values("Координата (правка)")
```

## Удалим данные с невалидными id

```python
df_brigada = df_brigada.drop(df_brigada[df_brigada["ID Парковки"] < 0].index)
df_inspec = df_inspec.drop(df_inspec[df_inspec["ID Парковки"] < 0].index)
```

# Одинаковые координаты у валидных парковок

```python
def brig_unique_cord(s):
    if len(df_brigada[df_brigada["Координата"] == s]["ID Парковки"].unique()) != 1:
        return True
    else:
        return False
def inspec_unique_cord(s):
    if len(df_inspec[df_inspec["Координата (правка)"] == s]["ID Парковки"].unique()) != 1:
        return True
    else:
        return False
```

## Ответы бригады

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_brigada[df_brigada["Координата"].apply(brig_unique_cord)].sort_values("Координата"))
```

## Ответы инспекции

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_inspec[df_inspec["Координата (правка)"].apply(inspec_unique_cord)].sort_values("Координата (правка)"))
```

# Разные координаты у одних id парковки
координаты у одной парковки могут меняться, но перепроверить набор нужно, поскольку могут попадать неправильные данные

```python
def brig_unique_id(s):
    if len(df_brigada[df_brigada["ID Парковки"] == s]["Координата"].unique()) != 1:
        return True
    else:
        return False
def inspec_unique_id(s):
    if len(df_inspec[df_inspec["ID Парковки"] == s]["Координата (правка)"].unique()) != 1:
        return True
    else:
        return False
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_brigada[df_brigada["ID Парковки"].apply(brig_unique_id)].sort_values(["ID Парковки", "Координата"]))
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_inspec[df_inspec["ID Парковки"].apply(inspec_unique_id)].sort_values(["ID Парковки", "Координата (правка)"]))
```

## Количество строк в двух таблицах выше
Возможно нецелесообразно проверять их вручную

```python
print("Ответы бригады - ", len(df_brigada[df_brigada["ID Парковки"].apply(brig_unique_id)]))
print("Ответы инспекции - ", len(df_inspec[df_inspec["ID Парковки"].apply(inspec_unique_id)]))
```

## Существует дисбаланс между изменением координат в ответах бригады и ответах инспекции почти в два раза
Это может намекать на то, что в данных проявляются ошибки<br>
ниже количество уникальных id парковок, у которых были изменены координаты

```python
print("Ответы бригады - ", len(df_brigada[df_brigada["ID Парковки"].apply(brig_unique_id)]["ID Парковки"].unique()))
print("Ответы инспекции - ", len(df_inspec[df_inspec["ID Парковки"].apply(inspec_unique_id)]["ID Парковки"].unique()))
```
